```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import powerlaw
```

```python
mobydick = pd.read_csv('mobydick.csv')
terror = pd.read_csv('terror.csv')
blackout = pd.read_csv('blackout.csv')
flare = pd.read_csv('flare.csv')
population = pd.read_csv('population.csv')
wildfire = pd.read_csv('wildfire.csv')
```

```python
pl_mobydick = powerlaw.Fit(mobydick['word'])
pl_terror = powerlaw.Fit(terror['terror'])
pl_blackout = powerlaw.Fit(blackout['blackout'])
pl_flare = powerlaw.Fit(flare['flare'])
pl_population = powerlaw.Fit(population['population'])
pl_wildfire = powerlaw.Fit(wildfire['wildfire'])
```

```python
print('',pl_mobydick.alpha, pl_mobydick.xmin, "\n",
      pl_terror.alpha, pl_terror.xmin, '\n',
      pl_blackout.alpha, pl_blackout.xmin, '\n',
      pl_flare.alpha, pl_flare.xmin, '\n',
      pl_population.alpha, pl_population.xmin, '\n',
      pl_wildfire.alpha, pl_wildfire.xmin, '\n',)
```

```python
mobydick_x, mobydick_y = pl_mobydick.cdf()
terror_x, terror_y = pl_terror.cdf()
blackout_x, blackout_y = pl_terror.cdf()
flare_x, flare_y = pl_terror.cdf()
population_x, populatoin_y = pl_terror.cdf()
wildfire_x, wildfire_y = pl_terror.cdf()
```

```python
fig1 = pl_mobydick.plot_ccdf(color = 'r')
pl_mobydick.power_law.plot_ccdf(color = 'r', linestyle='--', ax=fig1)

pl_mobydick.plot_pdf(color = 'g')
pl_mobydick.power_law.plot_pdf(color = 'g', linestyle='--', ax=fig1)
```

```python
plt.plot(pl_mobydick.xmins, pl_mobydick.Ds)
plt.show()

plt.plot(pl_mobydick.xmins, pl_mobydick.Ds, 'r')
plt.xlim(0,20)

index_mobydick = np.where(pl_mobydick.xmins == pl_mobydick.xmin)
plt.plot(pl_mobydick.xmin, pl_mobydick.Ds[index_mobydick],'o', color='lime')

plt.show()

print(pl_terror.xmin)
```
