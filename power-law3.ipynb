```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fl=pd.read_csv('flare.csv')
```

```python
fl=fl[fl['177']>170]
```

```python
def slope(x):
    le = len(x)
    m = min(x)
    s = 0
    for i in x:
        s += np.log(i/m)
    return 1+le*1/s
```

```python
slope(fl['177'])
```
