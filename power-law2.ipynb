```python
name_f = pd.read_csv('name_f.csv')
name_m = pd.read_csv('name_m.csv')
alpha = pd.concat([name_f, name_m])
alpha = alpha.sort_values(by=['이름수'], axis = 0)

flare = pd.read_csv('flare.csv')
flare = flare.sort_values(by=['177'], axis=0)
```

```python
def slope_m(x, m):
    sum = 0
    le = len(x>m)
    a = m-0.5
    for i in x:
        sum += np.log(i/a)
    return 1+le*1/sum
```

```python
def p_l(x):
    a,x_a,_ = plt.hist(x, 
                       bins=np.logspace(np.log10(10),np.log10(max(x)), 100), 
                       histtype='step')
    
    cum = np.cumsum(a)
    
    return x_a[1:], len(x)-cum
```

```python
a,x_a,_ = plt.hist(alpha['이름수'], 
                   bins=np.logspace(np.log10(10),np.log10(40000), 50), 
                   histtype='step')
```

```python
cum = np.cumsum(a)
```

```python
plt.plot(x_a[:-1],len(alpha)-cum)
```

```python
a_centers = 0.5*(x_a[1:]+x_a[:-1])
plt.plot(a_centers,len(alpha)-cum,'o')

plt.xscale("log")
plt.yscale("log")
plt.show()
```
